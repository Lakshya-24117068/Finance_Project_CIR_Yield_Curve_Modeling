# Notebook 03: CIR Parameter Calibration
We calibrate the CIR model parameters using OLS, MLE (exact likelihood), and GMM, and check the Feller boundary condition.


```python
import sys
import os
sys.path.insert(0, os.path.abspath('..'))
import numpy as np
from src.data.loader import DataLoader
from src.calibration.ols import OLSCalibrator
from src.calibration.mle import MLECalibrator
from src.calibration.gmm import GMMCalibrator

loader = DataLoader()
train_df = loader.load_processed_data('train_data.csv')
rt = train_df['ZC025YR'].values

print("Running OLS Calibration...")
ols = OLSCalibrator()
k_ols, th_ols, sig_ols = ols.calibrate(rt)
print(f"OLS: kappa={k_ols:.4f}, theta={th_ols:.4f}, sigma={sig_ols:.4f}")

print("\nRunning MLE Calibration...")
mle = MLECalibrator()
k_mle, th_mle, sig_mle = mle.calibrate(rt)
print(f"MLE: kappa={k_mle:.4f}, theta={th_mle:.4f}, sigma={sig_mle:.4f}")

print("\nRunning GMM Calibration...")
gmm = GMMCalibrator()
k_gmm, th_gmm, sig_gmm = gmm.calibrate(rt)
print(f"GMM: kappa={k_gmm:.4f}, theta={th_gmm:.4f}, sigma={sig_gmm:.4f}")
```
