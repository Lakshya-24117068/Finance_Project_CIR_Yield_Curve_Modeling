# Notebook 02: Exploratory Yield Curve Analysis
We plot yield curves over time and explore term structure changes between the training set (2016-2024) and the test set (2024-2026).


```python
import sys
import os
sys.path.insert(0, os.path.abspath('..'))
import matplotlib.pyplot as plt
import pandas as pd
from src.data.loader import DataLoader
from src.utils.constants import MATURITIES_MAP

loader = DataLoader()
train_df = loader.load_processed_data('train_data.csv')
test_df = loader.load_processed_data('test_data.csv')

avg_train = train_df[list(MATURITIES_MAP.keys())].mean()
avg_test = test_df[[c for c in MATURITIES_MAP.keys() if c in test_df.columns]].mean()

plt.figure(figsize=(10, 5))
plt.plot(list(MATURITIES_MAP.values()), avg_train.values*100, 'o-', label='Train Average')
plt.plot([MATURITIES_MAP[c] for c in avg_test.index], avg_test.values*100, 's-', label='Test Average (Inverted Curve)')
plt.title('Yield Curve Term Structure')
plt.xlabel('Maturity (Years)')
plt.ylabel('Yield (%)')
plt.legend()
plt.show()
```
