# Notebook 01: Data Engineering and Preprocessing
This notebook loads the raw yield curve data, removes duplicates, handles whitespace issues, and performs outlier smoothing.


```python
import sys
import os
sys.path.insert(0, os.path.abspath('..'))

import pandas as pd
from src.data.loader import DataLoader
from src.data.cleaner import DataCleaner
from src.data.validation import DataValidator

loader = DataLoader()
cleaner = DataCleaner()
validator = DataValidator()

print("Loading train data...")
train_raw = loader.load_raw_data('train_data.csv')
print(f"Raw Train Shape: {train_raw.shape}")

cleaned_train = cleaner.clean(train_raw)
print(f"Cleaned Train Shape: {cleaned_train.shape}")

# Save cleaned data
loader.save_processed_data(cleaned_train, 'train_data.csv')

# Validate
diag = validator.validate(cleaned_train, cleaned_train.columns)
print("Train Data Validation passed:", diag['valid'])
```
